## Algorito Apriori

```r
#Definición de parámetros iniciales
min_support <- 0.05
min_conf    <- 0.50
```

### Relacionado a Y1 (Heating Load)

```r
dsY1$Y1 <- as.factor(dsY1$Y1)
rules <- apriori( 
            dsY1, 
            parameter = list( support = min_support, conf = min_conf, minlen = 2, maxlen = 7 ),
            appearance = list( rhs = c("Y1=HL1", "Y1=HL2", "Y1=HL3", "Y1=HL4", "Y1=HL5") )
        )

inspect(rules)
```

```r
# Resumen de las reglas
summary( rules )
```

```r
# Se revisa si existen reglas duplicadas
duplicated( rules )
```

```r
# Se revisa si existen reglas redundantes
redundant <- is.redundant( rules )
redundant
which( is.redundant( rules ) )
```

```r
# Como existen, se utilizan sólo reglas no redundantes
rules <- rules[!redundant]
inspect( rules )
```

```r
# Resumen de apriori filtrado sin redundantes
summary( rules )
```

### Indicadores

```r
calculate_metrics <- function(cm) {
  confusion_matrix <- as.matrix(cm$table)
  
correct_predictions <- sum(diag(confusion_matrix))
  total_predictions <- sum(confusion_matrix)
  accuracy <- correct_predictions / total_predictions
  
  sensitivity <- numeric()
  specificity <- numeric()
  precision <- numeric()
  
  for (i in 1:nrow(confusion_matrix)) {
    TP <- confusion_matrix[i, i]
    FN <- sum(confusion_matrix[i, ]) - TP
    FP <- sum(confusion_matrix[, i]) - TP
    TN <- sum(confusion_matrix) - TP - FN - FP
    
    sensitivity[i] <- ifelse((TP + FN) == 0, NA, TP / (TP + FN))
    specificity[i] <- ifelse((TN + FP) == 0, NA, TN / (TN + FP))
    precision[i] <- ifelse((TP + FP) == 0, NA, TP / (TP + FP))
  }
  
  metrics_by_class <- data.frame(
    Class = rownames(confusion_matrix),
    Sensitivity = sensitivity,
    Specificity = specificity,
    Precision = precision
  )
  
  metrics <- list(
    Accuracy = accuracy,
    MetricsByClass = metrics_by_class
  )
  
  return(metrics)
}
```

```r
metrics <- calculate_metrics(cm_1)
print(paste("Overall Accuracy: ", metrics$Accuracy))
print(metrics$MetricsByClass)
```

```r
metrics <- calculate_metrics(cm_2)
print(paste("Overall Accuracy: ", metrics$Accuracy))
print(metrics$MetricsByClass)
```

### Comparación entre clasificadores fuertes y débiles

#### Para Y1

```r
indexes <- createDataPartition(dsY1$Y1, p=0.80, list = F)
train <- dsY1[indexes,]
test <- dsY1[-indexes,]

classificator_strong <- CBA(
  Y1 ~ ., data = train, supp = 0.1, conf=0.8, verbose = FALSE
)

classificator_weak   <- CBA(
  Y1 ~ ., train, supp = 0.001, conf = 0.8, verbose = FALSE
)
```

```r
predicted_strong <- predict(classificator_strong, test)
predicted_weak <- predict(classificator_weak, test)

cross_tab_strong<- table(predicted = predicted_strong, true = test$Y1)
cross_tab_weak<- table(predicted = predicted_weak, true = test$Y1)

cross_tab_strong
cross_tab_weak

accuracy_strong <- (cross_tab_strong[1,1]+cross_tab_strong[2,2]+cross_tab_strong[3,3])/sum(cross_tab_strong)
accuracy_weak <- (cross_tab_weak[1,1]+cross_tab_weak[2,2]+cross_tab_weak[3,3])/sum(cross_tab_weak)
```

```r
paste(accuracy_strong, accuracy_weak)
```
