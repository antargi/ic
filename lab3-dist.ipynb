```r
dt1 <- read_delim("./ENB2012_data_f.csv")
dt2 <- read_delim("./DataDiscretizado.csv")

# Seleccionar las columnas de interés
dt1_selected <- dt1[, c("X1", "X2", "X3", "X4", "X5", "X6", "X7", "X8")]
dt2_selected <- dt2[, c("Y1", "Y2")]

# Unir los datasets
merged_data <- cbind(dt1_selected, dt2_selected)
```

```r
boxplot.X1 = ggboxplot( data = merged_data, x = "Y1", y = "X1", color = "Y1", add = "jitter", shape = "Y1" ) + border() + ggtitle("Gráfico distribución de X1 por clase Y1 ")
ydens = axis_canvas( boxplot.X1, axis = "y", coord_flip = TRUE ) + geom_density( data = merged_data, aes( x = X1, fill = Y1), alpha = 0.7, linewidth = 0.2 ) + coord_flip()
boxplot.X1 = insert_yaxis_grob( boxplot.X1, ydens, grid::unit( .2, "null"), position = "right" )
ggdraw( boxplot.X1 )
```
